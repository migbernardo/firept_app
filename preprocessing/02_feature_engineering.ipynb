```python
data.iloc[data[data['category'] == 'desconhecida'].index, 13] = 'unknown' 
data.iloc[data[data['category'] == 'uso do fogo'].index, 13] = 'fire use' 
data.iloc[data[data['category'] == 'incendiarismo'].index, 13] = 'arsonism' 
data.iloc[data[data['category'] == 'reacendimento'].index, 13] = 'rekindling'  
data.iloc[data[data['category'] == 'acidentais'].index, 13] = 'accidental'  
data.iloc[data[data['category'] == 'estruturais'].index, 13] = 'structural' 
```

```python
data['negligent'] = 0 
data['intentional'] = 0 
data['negligent'] = data.iloc[data[data['to_del'] == 'negligente'].index, 11] 
data['intentional'] = data.iloc[data[data['to_del'] == 'intencional'].index, 11] 
data.iloc[data[data['negligent'].isna()].index, -2] = 0 
data.iloc[data[data['negligent'] == 'negligente'].index, -2] = 1 
data.iloc[data[data['intentional'].isna()].index, -1] = 0 
data.iloc[data[data['intentional'] == 'intencional'].index, -1] = 1 
data.iloc[data[data['rekindling'].isna()].index, 9] = 0 
data['rekindling'] = data['rekindling'].astype('int32')
data.drop(['code_cause', 'to_del'], axis=1, inplace=True)
```

```python
def null_datetime(x): 
    if str(x) != 'nan': 
        return dt.datetime.strptime(str(x), '%Y-%m-%d %H:%M:%S')
    else: 
        return x
```

```python
data['start_datetime'] = data['start_date'] +' '+ data['start_time'] 
data['end_datetime'] = data['end_date'] +' '+ data['end_time'] 
data['start_datetime'] = data['start_datetime'].apply(lambda x: null_datetime(x)) 
data['end_datetime'] = data['end_datetime'].apply(lambda x: null_datetime(x)) 
data['extinction_time'] = data['end_datetime'] - data['start_datetime'] 
data.drop(['start_date', 'start_time', 'end_date', 'end_time', 'extinction_days'], axis=1, inplace=True)
```

```python
def null_month(x): 
    if str(x) != 'NaT': 
        return x.strftime('%b')
    else: 
        return x
```

```python
data['month'] = data['start_datetime'].apply(lambda x: null_month(x))
```
